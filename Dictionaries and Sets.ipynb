# Dictionary
Dictionaries are a type of collection in Python. 

A dictionary has keys and values. 

The key is analogous to the index. They are like addresses, but thet don't have to be intergers. They are usually characters.



* Dictionaries are denoted with curly Brackets {}
* The keys have to be immutable and unique
* The valuew can be immutable, mutable and duplicates
* Each key and value pair is separate by a comma



# Sets

* Sets are a type of collection. This means that like lists and tuples you can input different Python types
* Unlike lists and tuples they are unordered. This means sets do not record element position
* Sets only have unqiue elements. This means there is only one of a particular element ina set



## Creating a Set

```python
album_list = ["Michael Jackson", "Thriller", "Thriller", 1982]

album_set = set(album_list)

album_set
```

We see that duplicate element does not exist

What about intersection?

```python
album_set_1 = {"AC/DC", "Back in Black", "Thriller"}
album_set_2 = {"AC/DC", "Back in Black", "The Dark Side of the Moon"}

album_set_3 = album_set_1 & album_set_2
album_set_3
```

What about union?

```python
album_set_1 = {"AC/DC", "Back in Black", "Thriller"}
album_set_2 = {"AC/DC", "Back in Black", "The Dark Side of the Moon"}

album_set_4 = album_set_1.union(album_set_2)
album_set_4
```
